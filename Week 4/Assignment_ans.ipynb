```python
inner_join = df.merge(df2, on='No.', how='inner')
outer_join = df.merge(df2, on='No.', how='outer')
left_join = df.merge(df2, on='No.', how='left')
right_join = df.merge(df2, on='No.', how='right')

print("\nInner Join Result:")
print(inner_join)

print("\nOuter Join Result:")
print(outer_join)

print("\nLeft Join Result:")
print(left_join)

print("\nRight Join Result:")
print(right_join)
```

```python
# Applying a function to a column
df['Age_Squared'] = df['Age'].apply(lambda x: x**2 if pd.notna(x) else x)
# Applying a function element- wise
df = df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
```

```python
filtered_df = df[df['Hobbies'].isin(['Music', 'Video Games'])]
print(filtered_df)
```

```python
# this is to show the median ages for both dataframes

grouped_df = df.groupby('Almamater')['Age'].mean().reset_index()
print("Old_Persona")
print(grouped_df)

grouped_df2 = df2.groupby('Almamater')['Age'].mean().reset_index()
print("New_persona")
print(grouped_df2)
```

```python
school_count = df.groupby('Almamater')['No.'].count().reset_index()
print("Old_Persona")
print(school_count)

school_count = df2.groupby('Almamater')['No.'].count().reset_index()
print("New_persona")
print(school_count)
```
